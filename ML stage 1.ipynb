```python
from sklearn.linear_model import LinearRegression
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
X = df3.drop(columns='Research & development expenses\nth EUR 2023')  # 34 features
y = df3['Research & development expenses\nth EUR 2023']               # Target variable

```

```python
reg = LinearRegression().fit(X, y)
reg.score(X, y)
```
